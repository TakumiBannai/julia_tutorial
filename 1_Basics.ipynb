```julia
# 可視化
using Plots
```

```julia
plot(
    x->x^3+x^2+x,       # ->で無名関数を指定
    xlabel="x",         # x軸凡例
    ylabel="y",         # y軸凡例
    label="y=x³+x²+x",  # グラフラベル
    xlims=(-3, 3),      # x軸範囲
    ylims=(-10, 10),    # y軸範囲
    xticks=-3:0.5:3,    # x軸目盛り表示刻み  
    title="Plot()",     # グラフタイトル
    linecolor=:blue,  # 色
    linewidth=3,        # 太さ
    linestyle=:dot,     # 描画スタイル
)
```

```julia
#plot(データ配列,st=:グラフの種類,...) stはseriestypeのAlias
p1=plot(rand(50),st=:bar,title="bar")                           #棒グラフ
p2=plot(rand(5),st=:pie,title="pie")                            #円グラフ
x=0:0.2:10
p3=plot(x,exp.(-x),st=:scatter,title="scatter",label="exp(-x)") #散布図 plot(xデータ,yデータ,st=...)
p4=plot(rand(1000)+rand(1000)+rand(1000)+rand(1000),
        st=:histogram,title="histogram")                        #ヒストグラム(4回乱数を足したデータを自動処理)
plot(p1,p2,p3,p4,layout=(1,4),size=(1000,300))                  #上記プロットを一気に1×4レイアウトでプロット
```

```julia
#データ作成
x=-3:0.2:3
y=x
z=@. exp(-(x^2+y'^2))                                    #z=exp(-(x²+y²)) , yの[']転置と@.マクロに注意
#各プロット
p1=heatmap(z,title="heatmap")                            #ヒートマップ heatmap(2次元データ,...)
p2=contour(x,y,z,title="contour")                        #コンター図   contour(xデータ,y,z,...)
p3=wireframe(x,y,z,title="wireframe")                    #3Dのワイヤーフレーム表記
p4=surface(x,y,z,title="surface")                        #3Dの曲面表示
plot(p1,p2,p3,p4,layout=(1,4),size=(1500,250),fmt=:png)  #1×4表記 (少し重たくなるので、後述のフォーマットpng)
```

```julia
# CSVデータ読み込み
using CSV
using DataFrames
```

```julia
# データの読み込み
boston = CSV.read("./data/Boston.csv", DataFrame)
```

```julia
# 基本統計量
describe(boston)
```

```julia
# Dataframe要素へのアクセス
age = boston.age
```

```julia
# 基本統計量
using Statistics
```

```julia
println("Mean: ", Statistics.mean(age))
println("Std: ", Statistics.std(age))
println("Median: ", Statistics.median(age))
println("Var: ", Statistics.var(age))
println("Max: ", maximum(age))
println("Min: ", minimum(age))
```

```julia
# ヒストグラム
using StatsPlots
```

```julia
StatsPlots.histogram(age, bins=20)
```
